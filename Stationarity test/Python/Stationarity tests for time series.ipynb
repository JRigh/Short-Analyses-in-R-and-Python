### Stationarity tests for time series

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2024)

# parameters
n = 100                 # Number of time points
phi = 0.7               # AR(1) coefficient for stationary series
sigma = 1               # Standard deviation of noise

# stationary time series: AR(1)
t1 = np.zeros(n)
epsilon = np.random.normal(0, sigma, n)  # White noise
for t in range(1, n):
    t1[t] = phi * t1[t - 1] + epsilon[t]

# Non-stationary time series: Random walk
t2 = np.zeros(n)
epsilon = np.random.normal(0, sigma, n)  # White noise
for t in range(1, n):
    t2[t] = t2[t - 1] + epsilon[t]

    # Plot the two series
plt.figure(figsize=(10, 6))

# Plot stationary time series
plt.subplot(2, 1, 1)
plt.plot(t1, color='darkred')
plt.title("Stationary Time Series (AR(1))")
plt.ylabel("Value")
plt.xlabel("Time")

# Plot non-stationary time series
plt.subplot(2, 1, 2)
plt.plot(t2, color='red')
plt.title("Non-Stationary Time Series (Random Walk)")
plt.ylabel("Value")
plt.xlabel("Time")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss

# Augmented Dickey-Fuller (ADF) Test
adf_result = adfuller(t1)
print("Augmented Dickey-Fuller Test:")
print(f"ADF Statistic: {adf_result[0]}")
print(f"p-value: {adf_result[1]}")
    
adf_result = adfuller(t2)
print("Augmented Dickey-Fuller Test:")
print(f"ADF Statistic: {adf_result[0]}")
print(f"p-value: {adf_result[1]}")
```

```python
# KPSS Test
kpss_result = kpss(t1, regression='c')  # 'c' for constant, 'ct' for trend
print("\nKPSS Test:")
print(f"KPSS Statistic: {kpss_result[0]}")
print(f"p-value: {kpss_result[1]}")
    
kpss_result = kpss(t2, regression='c')  # 'c' for constant, 'ct' for trend
print("\nKPSS Test:")
print(f"KPSS Statistic: {kpss_result[0]}")
print(f"p-value: {kpss_result[1]}")
```
